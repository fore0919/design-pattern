## Decorator Pattern

### 데코레이터 패턴이란? 
****
- 데코레이터 패턴(Decorator Pattern)은 객체 지향 디자인 패턴 중 하나로, 객체에 추가적인 기능을 동적으로 더할 수 있게 해주는 패턴.
- 이 패턴은 기존 객체를 수정하지 않고도 객체의 기능을 확장할 수 있다.
- 생성자에 들어갈 매개 변수를 메서드로 하나하나 받아들이고 마지막에 통합 빌드해서 객체를 생성하는 방식
- 복잡한 객체의 생성 알고리즘과 조립 방법을 분리하여 빌드 공정을 구축하는것이 목적

</br> 

### Wrapper
****

데코레이터 패턴의 핵심 아이디어는 객체를 감싸는 래퍼(wrapper)를 생성하여 이 래퍼 객체가 추가적인 기능을 제공하도록 하는 것이다.
래퍼는 기존 객체와 같은 인터페이스를 가지며, 기존 객체와 동일한 기능을 제공하며 래퍼 객체는 기존 객체에 덧붙여져서 새로운 객체를 형성.

</br> 

### 장점
****
1. 개방/폐쇄 원칙(OCP)을 따르기 쉽다. 
    기존 코드를 수정하지 않고도 새로운 기능을 추가할 수 있으므로, 코드의 수정이나 변경에 유연하게 대처 가능.
2. 객체 지향 디자인 원칙(SOLID)을 따른다. 
    래퍼 객체를 통해 코드의 중복을 제거하고 객체 간의 결합도를 낮추어 유지보수성과 확장성을 향상.
3. 기존 객체에 영향을 미치지 않으므로, 코드의 안정성과 신뢰성을 보장.
    높은 수준의 재사용성을 달성할 수 있습니다.

</br> 

### 단점
****
1. 객체의 수가 증가할 수 있습니다 -> 데코레이터를 많이 사용하면 성능에 영향을 미칠 수 있다.
2. 래퍼 객체를 생성할 때 추가적인 비용이 들어감.


```python
import time

def time_it(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Elapsed time: {end - start}")
        return result
    return wrapper

@time_it
def my_function(n):
    return sum(range(n))

result = my_function(1000000)
print(result)
```
